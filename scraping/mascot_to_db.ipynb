```python
# Read the csv file for the data
csv_path = '../website/static/Resources/team_mascot_2018.csv'
team_df = pd.read_csv(csv_path)
team_df.head()
```

```python
# Rename the columns
team_df.columns = ['region', 'seed', 'name', 'mascot']
team_df.head()
```

```python
new_df = team_df.sort_values('region')
new_df.head()
```

```python
east_df = new_df.iloc[0:18]
drop_east_df = east_df.drop(['region'], axis=1)
new_east_df = drop_east_df.sort_values('seed')

new_east_df = new_east_df.reset_index(drop=True)
new_east_df
```

```python
midwest_df = new_df.iloc[18:35]
drop_midwest_df = midwest_df.drop(['region'], axis=1)
new_midwest_df = drop_midwest_df.sort_values('seed')
new_midwest_df = new_midwest_df.reset_index(drop=True)
new_midwest_df
```

```python
south_df = new_df.iloc[35:51]
drop_south_df = south_df.drop(['region'], axis=1)
new_south_df = drop_south_df.sort_values('seed')
new_south_df = new_south_df.reset_index(drop=True)
new_south_df
```

```python
west_df = new_df.iloc[51:68]
drop_west_df = west_df.drop(['region'], axis=1)
new_west_df = drop_west_df.sort_values('seed')
new_west_df = new_west_df.reset_index(drop=True)
new_west_df
```

```python
rec_east = json.loads(new_east_df.T.to_json()).values()
rec_midwest = json.loads(new_midwest_df.T.to_json()).values()
rec_south = json.loads(new_south_df.T.to_json()).values()
rec_west = json.loads(new_west_df.T.to_json()).values()
```

```python
east_json = dumps(rec_east)

with open('../website/static/resources/east_teams.json', 'w') as jf:
    jf.write(east_json)
```

```python
midwest_json = dumps(rec_midwest)

with open('../website/static/resources/midwest_teams.json', 'w') as jf:
    jf.write(midwest_json)
```

```python
south_json = dumps(rec_south)

with open('../website/static/resources/south_teams.json', 'w') as jf:
    jf.write(south_json)
```

```python
west_json = dumps(rec_west)

with open('../website/static/resources/west_teams.json', 'w') as jf:
    jf.write(west_json)
```
